To samo ale wykres punktowy

```python
plt.plot(df['math score'], "*", label = 'math', color="salmon")
plt.plot(df['reading score'], "^", label = 'reading', color="sienna")
plt.plot(df['writing score'], ".", label = 'writing', color="cornflowerblue")
plt.xlabel('Sample')
plt.ylabel('Test score')
plt.legend()
```

subplot powoduje Å¼e na jednym rysunku mamy narysowane 3 rÃ³Å¼ne wykresy

```python
plt.subplot(3, 1, 1)
#1 liczba - ile wierszy wykresÃ³w bÄ™dzie
#2 liczba - ile kolumn wykresÃ³w bÄ™dzie
#3 liczba - ktÃ³ry wykres definiujemy
plt.plot(df['math score'], "*", label = 'math', color="salmon")
plt.ylabel('Test score')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df['reading score'], "^", label = 'reading', color="sienna")
plt.ylabel('Test score')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df['writing score'], ".", label = 'writing', color="cornflowerblue")
plt.xlabel('Sample')
plt.ylabel('Test score')
plt.legend()
```

```python
plt.subplot(1, 3, 1)
plt.plot(df['math score'], "*", label = 'math', color="salmon")
plt.xlabel('Sample')
plt.ylabel('Test score')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(df['reading score'], "^", label = 'reading', color="sienna")
plt.xlabel('Sample')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(df['writing score'], ".", label = 'writing', color="cornflowerblue")
plt.xlabel('Sample')
plt.legend()
plt.savefig('costam.png')
```
