Prosty wykres

```python
import matplotlib.pyplot as plt
plt.plot(df['math score'])
```

Już na wstępie wiem że coś jest nie tak z tym wykresem,

Zadaj sobie pytanie czy: osoba czytająca zrozumie co on przedstawia i czy typ wykresu jest dobrze dobrany

```python
plt.plot(df['math score'])
plt.xlabel('Sample')
plt.ylabel('Test score')
```

```python
#dorzucamy inne dane i legendę
plt.plot(df['math score'], label = 'math')
plt.plot(df['reading score'], label = 'reading')
plt.plot(df['writing score'], label = 'writing')
plt.xlabel('Sample')
plt.ylabel('Test score')
plt.legend()
```

To samo ale wykres punktowy

```python
plt.plot(df['math score'], "*", label = 'math', color="salmon")
plt.plot(df['reading score'], "^", label = 'reading', color="sienna")
plt.plot(df['writing score'], ".", label = 'writing', color="cornflowerblue")
plt.xlabel('Sample')
plt.ylabel('Test score')
plt.legend()
```

subplot powoduje że na jednym rysunku mamy narysowane 3 różne wykresy

```python
plt.subplot(3, 1, 1)
#1 liczba - ile wierszy wykresów będzie
#2 liczba - ile kolumn wykresów będzie
#3 liczba - który wykres definiujemy
plt.plot(df['math score'], "*", label = 'math', color="salmon")
plt.ylabel('Test score')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df['reading score'], "^", label = 'reading', color="sienna")
plt.ylabel('Test score')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df['writing score'], ".", label = 'writing', color="cornflowerblue")
plt.xlabel('Sample')
plt.ylabel('Test score')
plt.legend()
```

```python
plt.subplot(1, 3, 1)
plt.plot(df['math score'], "*", label = 'math', color="salmon")
plt.xlabel('Sample')
plt.ylabel('Test score')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(df['reading score'], "^", label = 'reading', color="sienna")
plt.xlabel('Sample')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(df['writing score'], ".", label = 'writing', color="cornflowerblue")
plt.xlabel('Sample')
plt.legend()
plt.savefig('costam.png')
```
