główna wada algorytmu k-means: to ja ustalam ilosc sąsiadów

```python
import pandas as pd

df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
#do dalszej analizy bierzemy tylko kolumny od Age włącznie
X_train = df.iloc[:,2:]

#ucinam - chce od drugiej kolumny (age)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"])
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```

```python
#klasteryzacja za pomocą K-means i 4 klastrami
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
model = kmeans.fit(X_train)
labels = model.predict(X_train) 
# przy predict staramy się nadać każdej klasie jakiś label
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```

```python
#zrób to samo dla 7 klastrów i wyświetl wykres

#klasteryzacja za pomocą K-means i 7 klastrami
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=7)
model = kmeans.fit(X_train)
labels = model.predict(X_train) 

# wykres
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```
